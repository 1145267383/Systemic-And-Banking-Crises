## Introduction

>**Systemic And Banking Crisis** This data contains the Systemic and the banking crises of 70 countries, from 1800 AD to 2016 AD, with a total of 8642 records and 17 variables.


### **Variable      |   Description***:
- **1-Case**: ID Number for Country.
- **2-Cc3**: ID String for Country.
- **3-Country**: Name Country.
- **4-Yea**: The date  from 1800 to 2016.
- **5-Banking_Crisis**: Banking problems can often be traced to a decrease the value of banks' assets:
>- A) due to a collapse in real estate prices or When the bank asset  values decrease substantially.
>- B) if a government stops paying its obligations, this can trigger a sharp decline in value of bonds.
- **6-Systemic_Crisis**: When many banks in a country are in serious solvency or liquidity problems at the same timeâ€”either:
>- A) because there are all hits by the same outside shock.
>- B) or because failure in one bank or a group of banks spreads to other banks in the system.

- **7-Gold_Standard**: The Country have crisis in Gold Standard.
- **8-Exch_Usd**: Exch local currency in USD, Except exch USD currency in GBP.
- **9-Domestic_Debt_In_Default**: The Country have domestic debt in default.
- **10-Sovereign_External_Debt_1**: Default and Restructurings, 1800-2012--Does not include defaults on WWI debt to United States                         and United Kingdom and post-1975 defaults on Official External Creditors.
- **11-Sovereign_External_Debt_2**: Default and Restructurings, 1800-2012--Does not include defaults on WWI debt to United                               States and United Kingdom but includes post-1975 defaults on Official External Creditors.
- **12-Gdp_Weighted_Default**: GDP Weighted Default for country.
- **13-Inflation**: Annual percentages of average consumer prices.
- **14-Independence**: Independence for country.
- **15-Currency_Crises**: The Country have crisis in Currency.
- **16-Inflation_Crises**: The Country have crisis in Inflation.
- **17-Level_Country**: The description of the country  whether it's Developing or Developed.

```python
def CountPercent(df,x,fig=[7,3]):
    g=sns.countplot(x=df[x])
    ttl=df[x].count()
    for p in g.patches:
            g.annotate(f'{str(round(p.get_height()/ttl*100, 2))}%', (p.get_x()+ 0.5*(p.get_width()), p.get_height()), 
                ha="center",  va="bottom",
                fontsize=14, weight="bold")
    sns.set(rc={'figure.figsize': list(fig)}, font_scale=1.2);
```

```python
import numpy as np
import pandas as pd 
import seaborn as sns 
import plotly
import plotly.io as pio
import plotly.express as px
import matplotlib.pyplot as plt 
from pandas.api.types import CategoricalDtype as catdt
%matplotlib inline

pio.renderers.default = 'browser'
sns.set(rc={'figure.figsize': [15,5]}, font_scale=1.2);
```

# view Data
#### `Sys_Ban_Crisis`
>- `not` Systemic_Crisis And with `not` Banking_Crisis >>>`N`
>- `not` Systemic_Crisis And with Banking_Crisis >>>`B`
>- Systemic_Crisis And with `not` Banking_Crisis >>>`S`
>- Systemic_Crisis And with Banking_Crisis >>>`SB`

```python
s_b_c=[]
for i,l in zip(df.Systemic_Crisis,df.Banking_Crisis):
    if i==1 and l==1:     #Systemic With Banking  >>>3
        x="SB"
    elif i==1 and l==0:  # Systemic With not Banking >>>2
        x="S"
    elif i==0 and l==1:  # not Systemic With Banking >>>1
        x="B"
    elif i==0 and l==0:
        x="N"             # not Systemic With not Banking >>>0
    s_b_c.append(x)

df["Sys_Ban_Crisis"]=np.array(s_b_c)

cat_dtype = catdt(categories=['N','B','S', 'SB'], ordered=True)
df["Sys_Ban_Crisis"]=df["Sys_Ban_Crisis"].astype(cat_dtype)

df=df.set_index("Year")
```

```python
CountPercent(df,"Sys_Ban_Crisis",fig=[12, 3])

n,sb,b,s=df["Sys_Ban_Crisis"].value_counts()
print(f""">- not` Systemic_Crisis And with `not` Banking_Crisis >>>`N`= {n}
>- `not` Systemic_Crisis And with Banking_Crisis >>>`B`= {b}
>- Systemic_Crisis And with `not` Banking_Crisis >>>`S`= {s}
>- Systemic_Crisis And with Banking_Crisis >>>`SB`= {sb}""")
```

```python
df["Sys_Ban_Crisis"]=df["Sys_Ban_Crisis"].astype("str")
df["Systemic_Crisis"]=df["Systemic_Crisis"].astype("str")
df['Banking_Crisis']=df['Banking_Crisis'].astype("str")
df['Gold_Standard']=df['Gold_Standard'].astype("str")
df['Domestic_Debt_In_Default']=df['Domestic_Debt_In_Default'].astype("str") 
df['Sovereign_External_Debt_1']=df['Sovereign_External_Debt_1'].astype("str") 
df['Sovereign_External_Debt_2']=df['Sovereign_External_Debt_2'].astype("str")
df['Independence']=df['Independence'].astype("str")
df['Currency_Crises']=df['Currency_Crises'].astype("str")
df['Inflation_Crises']=df['Inflation_Crises'].astype("str")
```

```python
fig = px.choropleth(df,
                    title="Systemic And Banking Crisis",
                    locations="Cc3",
                    projection="eckert3",
                    color="Sys_Ban_Crisis",
                    labels={"Sys_Ban_Crisis":"Systemic&Banking"},
                    color_discrete_map={"N":'Green', "B":'Yellow',"S":'Orange',"SB":'red'},
                    hover_name="Country",
                    animation_frame=df.index,
                    hover_data=["Systemic_Crisis", 'Banking_Crisis', 'Gold_Standard', 'Exch_Usd', 'Domestic_Debt_In_Default', 'Sovereign_External_Debt_1',"Sovereign_External_Debt_2", 'Gdp_Weighted_Default', 'Inflation', 'Independence', 'Currency_Crises', 'Inflation_Crises', 'Level_Country'])
```

```python
plotly.offline.plot(fig,filename="History.html",auto_open=True)
```
