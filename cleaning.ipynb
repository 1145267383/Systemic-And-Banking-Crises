```python
def Dropnan(df,c,ty=""):
    """
    Procnon:Processing of non values present in time series data.
    df:Name of Dataset.
    c:Name of columns.
    
    """
    dn=df[df[c].isnull()].index
    for n in dn:
        a=n-4;b=n+3
        if a<0:
            a=0;b=7
            
        dc=df.loc[a:b,c]
        dl=list(dc)
        d=list(dc.dropna())
        s=set(d)
        l=list(s)
            
        if ty=="str":
            c_l=[]
            for e in range(len(l)):
                c_l.append(d.count(l[e]))
                    
            dct=dict(zip(c_l,l)) 
            df.loc[n,c]=dct[np.max(c_l)]
            print(dl,l,c_l,dct[np.max(c_l)])
        else:
            df.loc[n,c]=float(round(np.mean(l),2))
            print(dl,l,round(np.mean(l),2))
                    
```
